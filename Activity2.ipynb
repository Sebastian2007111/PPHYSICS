```python
import matplotlib.pyplot as plt
import numpy as np

# Función para partículas aleatorias en 2D
def random_particles_2d(width, height, num_particles, particle_radius):
    positions = []
    while len(positions) < num_particles:
        x = np.random.uniform(particle_radius, width - particle_radius)
        y = np.random.uniform(particle_radius, height - particle_radius)
        new_position = np.array([x, y])
        if all(np.linalg.norm(new_position - pos) > 2 * particle_radius for pos in positions):
            positions.append(new_position)
    return np.array(positions)

# Función para partículas ordenadas en un cuadrado 2D (100x80)
def grid_particles_2d(width, height, num_particles):
    particles_per_row = int(np.sqrt(num_particles))
    x_spacing = width / (particles_per_row + 1)
    y_spacing = height / (particles_per_row + 1)
    
    positions = []
    for i in range(1, particles_per_row + 1):
        for j in range(1, particles_per_row + 1):
            if len(positions) < num_particles:
                positions.append([i * x_spacing, j * y_spacing])
    return np.array(positions)

# Parámetros para la caja 2D (aleatorio)
width, height = 100, 80
num_particles_2d = 20
particle_radius = 2

# Parámetros para el cuadrado 2D (ordenado)
square_width, square_height = 100, 80
num_particles_square = 20

# Generar partículas
particles_2d = random_particles_2d(width, height, num_particles_2d, particle_radius)
particles_square = grid_particles_2d(square_width, square_height, num_particles_square)

# Dibujar partículas aleatorias en 2D
fig, ax = plt.subplots(figsize=(8, 10))
ax.set_xlim(0, width)
ax.set_ylim(0, height)
ax.set_aspect('equal')
ax.set_title("Partículas aleatorias en 2D")
ax.set_xlabel("Ancho")
ax.set_ylabel("Altura")

for pos in particles_2d:
    circle = plt.Circle(pos, particle_radius, color='blue', alpha=0.7)
    ax.add_artist(circle)

rect = plt.Rectangle((0, 0), width, height, fill=None, edgecolor='black', linewidth=1.5)
ax.add_patch(rect)
plt.show()

# Dibujar partículas ordenadas en un cuadrado 2D
fig, ax = plt.subplots(figsize=(10, 8))
ax.set_xlim(0, square_width)
ax.set_ylim(0, square_height)
ax.set_aspect('equal')
ax.set_title("Partículas ordenadas en un cuadrado 2D (100x80)")
ax.set_xlabel("Ancho")
ax.set_ylabel("Altura")

for pos in particles_square:
    circle = plt.Circle(pos, particle_radius, color='red', alpha=0.7)
    ax.add_artist(circle)

rect = plt.Rectangle((0, 0), square_width, square_height, fill=None, edgecolor='black', linewidth=1.5)
ax.add_patch(rect)
plt.show()
```
